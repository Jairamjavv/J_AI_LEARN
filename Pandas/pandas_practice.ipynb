### Conditional selection

```python
# select price > 50
wine_reviews_df['price'] > 50
```

```python
# show all rows where price > 50
wine_reviews_df[wine_reviews_df['price'] > 50]
```

```python
# select only country and price where price > 50
wine_reviews_df.loc[wine_reviews_df['price'] > 50, ['country', 'price']]
```

```python
# select only country and price where price > 50 and country == France
wine_reviews_df.loc[(wine_reviews_df['country'] == 'France') & (wine_reviews_df['price'] > 50), ['country', 'price']]
```

```python
# select only country and price where price > 50 or country == France
wine_reviews_df.loc[(wine_reviews_df['country'] == 'France') | (wine_reviews_df['price'] > 50), ['country', 'price']]
```

```python
# select country and price for country in France and Spain
wine_reviews_df.loc[wine_reviews_df['country'].isin(['Spain', 'France']), ['country', 'price']]
```

```python
# check for null
wine_reviews_df.loc[wine_reviews_df['country'].isnull()]
```

```python
# check for not null
wine_reviews_df.loc[wine_reviews_df['country'].notnull()]
```
